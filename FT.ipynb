```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel("Fossil Fuels.xlsx")
```

```python
#sort values
df.sort_values(by=['%'], inplace=True)
```

```python
df['color'] = df['Continent'].map({'Europe':'#006f9b', 'Asia':'#f83', 'Africa':'#00994d', 'Others':'#262a33', 'North America':'#593380', 'South America':'#FF9900'})
```

```python
df.sort_values(by=['%'], ascending = False, inplace=True)
```

```python
plt.style.use("fivethirtyeight")
fig = plt.figure(figsize = (12, 8))
ax = fig.add_subplot(111)

ax.set_facecolor("#FFF0E5")
fig.set_facecolor("#FFF0E5")

#horizontal bar graph with color code of the color
#plot the bar chart and group by continent
ax = sns.barplot(x = '%', y = 'Country', data = df, palette = df['color'])

#outer border of the chart to be false
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

#show grid lines as rows and columns
ax.grid(True)

for p in ax.patches:
    ax.annotate(str(p.get_width()) + '%', (p.get_width() * 1.005, p.get_y() + 0.5), fontsize = 13)

#remove xticks and yticks
ax.set_xticks([])

#yticks font size
ax.set_yticklabels(ax.get_yticklabels(), fontsize = 12, fontname = 'Trebuchet MS')

#remove the x axis label
ax.set_xlabel('')
ax.set_ylabel('')

plt.title("Russia Exported $66 Billion of Fossil Fuels since the invasion of Ukraine", fontsize = 20, fontweight = 'bold', y = 1.05,  pad = 20, fontname = 'Trebuchet MS')
plt.suptitle("India imported 1.1% of these exports, Europe imported 70% and China 10%", fontsize = 16, y = 0.94, fontname = 'Trebuchet MS', alpha = 0.7)

#footer
plt.text(0.5, 22, 'Data for 2 months since Feb 24, 2022', fontsize = 12, ha = 'center', fontname = 'Trebuchet MS', alpha = 0.5)
plt.text(14, 22, 'Source: Centre for Research on Energy and Clean Air', fontsize = 12, ha = 'center', fontname = 'Trebuchet MS', alpha = 0.5)

plt.savefig('Fossil Fuels.png', dpi = 300, bbox_inches = 'tight', pad_inches = 0.1, facecolor = '#FFF0E5')
```
