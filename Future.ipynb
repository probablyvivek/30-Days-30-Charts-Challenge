```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("UN.csv")
```

```python
df.head()

#get average of Population(future projections and Population(historical estimates)
mean1 = df["Population (future projections)"].mean()
mean2 = df["Population (historical estimates)"].mean()

```

```python
#create an interface to get the estimates from input
def get_estimates(Entity):
    df_Entity = df[df["Entity"] == Entity]
    # print("The estimated population of {} is {}".format(Entity, df_Entity["Population (future projections)"].iloc[-1]))
    #plot the graph
    plt.style.use("fivethirtyeight")
    fig = plt.figure(figsize = (12, 8))
    ax = fig.add_subplot(111)
    fig.set_facecolor("#343a40")
    ax.set_facecolor("#343a40")
    ax.plot(df_Entity["Year"], df_Entity["Population (historical estimates)"], label = "Historical", color = "#0077b6", linestyle = "solid", linewidth = 2)
    ax.plot(df_Entity["Year"], df_Entity["Population (future projections)"], label = "Future", color = "#ff4d6d", linestyle = "dashed", linewidth = 2)
    ax.set_title("Population Projections by the UN, 1950-2100, {}".format(Entity), fontsize = 20, pad=20, fontname = "Trebuchet MS", color = "#ffffff")
    ax.set_xlabel("Year", fontsize = 15, color = "#ffffff")
    ax.set_ylabel("Population", fontsize = 15, color = "#ffffff")
    plt.grid(False)
    ax.grid(False)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.tick_params(axis = "x", colors = "#ffffff", labelsize = 15)
    ax.tick_params(axis = "y", colors = "#ffffff", labelsize = 15)
    #legend color
    legend = ax.legend(loc = "best", fontsize = 10, frameon = False)
    for text in legend.get_texts():
        text.set_color("#ffffff")
    plt.show()
    fig.savefig("UN_India.png", facecolor = "#343a40", edgecolor = "#343a40", transparent = True)
    return ax

    #call the function to get the estimates for each Entity
get_estimates("India")
```
