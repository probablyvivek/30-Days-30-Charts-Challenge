```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as stat
import plotly.express as px
```

```python
df = pd.read_csv('spotify.csv', encoding="ISO-8859-1")
```

```python
df.head()
```

```python
spotify_int = df.iloc[:, 4:17]
spotify_int.head()

```

```python
#drop mode 
spotify_int = spotify_int.drop(['mode'], axis=1)
spotify_int = spotify_int.drop(['duration_ms'], axis=1)
spotify_int = spotify_int.drop(['popularity'], axis=1)
spotify_int = spotify_int.drop(['instrumentalness'], axis=1)
spotify_int = spotify_int.drop(['key'], axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
sns.set(style="ticks", color_codes=True)
sns.set(style="darkgrid")

sns.pairplot(spotify_int)

#title
plt.title('Pairplot of Spotify Somgs of 2021')

#xlabel
plt.xlabel('Spotify Features')

#ylabel
plt.ylabel('Spotify Features')

#save
plt.savefig('spotify_pairplot.png')
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(14,12))
corr = spotify_int.corr()
heat = sns.heatmap(corr, vmax=1, square=True, annot=True, fmt='.2f', cmap='viridis')

#title
plt.title('Heatmap of Spotify Correlation')

#xlabel
```
